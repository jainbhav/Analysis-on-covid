```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('latest Covid-19 India Status1 (1).csv')
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.columns=['state','total_cases','active','deaths','active_ratio','death_ratio']
```

```python
df
```

```python
df.sort_values('active_ratio',ascending=False)
```

```python
plt.figure(figsize=(10,4))
plt.plot(df['state'],df['active_ratio'],color='blue',marker='o')
plt.xticks(rotation=90,fontsize=11)
plt.yticks(fontsize=13)
plt.ylabel('active ratio',fontsize=15)
plt.title('Line plot for active ratio')
plt.show()
```

# Total cases in each state

```python
total_cases_state=df.groupby(['state'])['total_cases'].max().sort_values(ascending=False).reset_index()
total_cases_state
```

```python
plt.figure(figsize=(15, 8))
plt.xticks(rotation=75)
sns.barplot(x='state',y='total_cases',data=total_cases_state)
plt.title('Total cases in each state')
plt.xticks(rotation=90,fontsize=15)
plt.xlabel('State')
plt.ylabel('Total cases')
plt.show()
```

```python
plt.plot(df['state'],df['active_ratio'],color='red',marker='o',label='active cases')
plt.plot(df['state'],df['death_ratio'],color='blue',marker='*',label='death')
plt.xticks(rotation=90,fontsize=10)
plt.legend()
plt.show()

            
```

# calculating recovered cases

```python
df['recovered']=df['total_cases']-(df['active']+df['deaths'])
df
```

```python
rec_cases=df.groupby(['state'])['recovered'].max().sort_values(ascending=False).reset_index()
rec_cases
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x='state',y='recovered',data=rec_cases)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x='state',y='recovered',data=rec_cases.head())
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.plot(df['state'],df['total_cases'],linestyle='--',color='red',label='total_cases')
plt.plot(df['state'],df['recovered'],color='yellow',marker='o',label='recover')
plt.plot(df['state'],df['deaths'],color='blue',marker='s',label='deaths')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
